```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pickle

plt.style.use("default")
sns.set_context("talk")
```

```python
plt.figure(figsize=(4.75, 4), dpi=150, facecolor="w")
plt.bar([0, 1], [25, 120], color=["#1E88E5", "#EDAC32"], edgecolor="k", linewidth=3)

plt.xlim(-0.5, 1.5)

plt.xticks([-0.5, 0, 1, 1.5], ["", "EGG", "GA", ""])

plt.ylabel("Generation time (s)")

sns.despine(trim=True)
```

```python
diff_within = [0.021, 0.016, 0.007, 0.011, 0.017, 0.018]
diff_cross = [0.098, 0.051, 0.050, 0.047, 0.039, 0.126]

ga_within = [0.057, 0.032, 0.013, 0.015, 0.025, 0.052]
ga_cross = [0.149, 0.084, 0.063, 0.156, 0.078, 0.223]
```

```python
plt.figure(figsize=(3.75, 3), dpi=150, facecolor="w")
plt.bar(
    [0 - 0.2, 1 - 0.2],
    [np.mean(diff_within), np.mean(ga_within)],
    color=["#1E88E5", "#1E88E5"],
    edgecolor="k",
    linewidth=3,
    width=0.4,
    label="EGG",
)
plt.bar(
    [0 + 0.2, 1 + 0.2],
    [np.mean(diff_cross), np.mean(ga_cross)],
    color=["#EDAC32", "#EDAC32"],
    edgecolor="k",
    linewidth=3,
    width=0.4,
    label="GA",
)

plt.xlim(-0.5, 1.5)

plt.xticks([-0.5, 0, 1, 1.5], ["", "Within", "Cross", ""])
plt.yticks([0, 0.05, 0.1, 0.15])

plt.ylabel("$L_2$ Distance")

sns.despine(trim=True)
plt.legend(frameon=False)
```
